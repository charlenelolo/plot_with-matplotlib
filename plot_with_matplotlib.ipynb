# Matplotlib

## Importing

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
y = 2*x**2
```

```python
a = np.linspace(0,10,11)
b = a**4
```

## plt.plot()

```python
plt.plot(x,y)
plt.xlabel('QTY')
plt.ylabel('price')
plt.title('Supply Curve')
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('QTY')
plt.ylabel('price')
plt.title('Supply Curve')
plt.xlim(0,9)
plt.ylim(20,100)
plt.show()
```

## Exporting a plot

```python
plt.plot(x,y)
plt.savefig('example.png')
```

## Creating a Figure

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
plt.show()
```

```python
fig = plt.figure(figsize=(12,8), dpi=100) #parameters

axes1 = fig.add_axes([0, 0, 1, 1]) #large figure
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25]) # Small figure
axes3 = fig.add_axes([1.2, 0.7, 0.25, 0.25]) # side figure

axes1.plot(a,b)
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Big Figure')

axes2.plot(a,b)
axes2.set_xlim(8,10)
axes2.set_ylim(4000,10000)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Zoomed In')

axes3.plot(a,b)
axes3.set_xlabel('X Label')
axes3.set_ylabel('Y Label')
axes3.set_title('Side Figure')

fig.savefig('figure.png', bbox_inches='tight')
```
